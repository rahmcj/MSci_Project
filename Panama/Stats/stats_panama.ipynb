```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'Latest_2808/chao1.csv'
df_chao1 = pd.read_csv(file_path, sep=',')


# Load the metadata file
metadata = pd.read_csv('metadata_grouped.tsv', sep='\t')

# Merge the Chao1 data with the metadata
merged_df = pd.merge(df_chao1, metadata[['sample-id', 'Depth_community']], on='sample-id', how='left')
```

```python
# Extract the columns related to depth from the merged DataFrame
depth_columns = [col for col in merged_df.columns if col.startswith('depth')]

# Extract unique depth values from column nam.es
depth_values = sorted(set(int(col.split('_')[0].split('-')[1]) for col in depth_columns))

# Create a new DataFrame with sample ID and mean Chao1 diversity for each depth
df_means = pd.DataFrame({
    'sample-id': merged_df['sample-id'],
    **{f'depth-{depth}': merged_df[[col for col in depth_columns if col.startswith(f'depth-{depth}')]].mean(axis=1) for depth in depth_values}
})

# Melt the DataFrame for easier plotting
df_melted = df_means.melt(id_vars=['sample-id'], var_name='Depth', value_name='Chao1_Diversity')

# Add the "Depth_community" from metadata
df_melted = pd.merge(df_melted, merged_df[['sample-id', 'Depth_community']], on='sample-id', how='left')

# Convert the 'Depth' column to numeric for plotting
df_melted['Depth'] = df_melted['Depth'].str.replace('depth-', '').astype(int)

# Plotting
plt.figure(figsize=(14, 8))

# Plot each sample's rarefaction curve, using "Depth_community" for labeling
for community in df_melted['Depth_community'].unique():
    community_data = df_melted[df_melted['Depth_community'] == community]
    plt.plot(community_data['Depth'], community_data['Chao1_Diversity'], marker='o', markersize=2, linewidth=1, label=community, alpha=0.7)

# Customize the plot
plt.xlabel('Sequencing Depth')
plt.ylabel('Chao1 Diversity Index')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', ncol=2, title='Depth Community')

plt.grid(True)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Calculate the mean Chao1 for each sample
depth_columns = [col for col in df_chao1.columns if col.startswith('depth')]
df_chao1['mean_chao1'] = df_chao1[depth_columns].mean(axis=1)

# Merge Chao1 data with metadata
merged_df = pd.merge(df_chao1[['sample-id', 'mean_chao1']], metadata[['sample-id', 'Phasic_community']], on='sample-id')

# Create the boxplot
plt.figure(figsize=(10, 8))
merged_df.boxplot(column='mean_chao1', by='Phasic_community', grid=False)

# Customize the plot
plt.title('Chao1 Diversity by Site')
plt.suptitle('')  # Remove the automatic title to clean up the plot
plt.xlabel('Site')
plt.ylabel('Mean Chao1 Index')

# Show the plot
plt.show()
```

```python
import pandas as pd

# Load the file contents
weighted_u = 'pcoa_results/wei_uni/ordination.txt'

# Read the entire file into a list of lines
with open(weighted_u, 'r') as f:
    lines = f.readlines()

# Split the sections based on header keywords
eigvals = []
proportion = []
coords = []

section = None
for line in lines:
    if line.startswith('Eigvals'):
        section = 'eigvals'
        continue
    elif line.startswith('Proportion explained'):
        section = 'proportion'
        continue
    elif line.startswith('Site'):
        section = 'coords'
        continue

    if section == 'eigvals':
        eigvals.append(line.strip())
    elif section == 'proportion':
        proportion.append(line.strip())
    elif section == 'coords':
        coords.append(line.strip())

# Convert the eigvals and proportion sections to lists of floats
eigvals = list(map(float, eigvals[0].split()))
proportion_explained = list(map(float, proportion[0].split()))

# Convert the coords section to a DataFrame
coords_df = pd.DataFrame(
    [x.split('\t')[1:] for x in coords[1:]],  # Skip the first header row
    columns=[f'PC{i+1}' for i in range(len(coords[0].split('\t'))-1)],  # -1 to account for the sample ID
    index=[x.split('\t')[0] for x in coords[1:]]
)

# Convert values to floats
coords_df = coords_df.astype(float)

# Display the DataFrame and eigenvalues
print(coords_df.head())
print(f"Eigenvalues: {eigvals}")
print(f"Proportion explained: {proportion_explained}")
```

```python
### Making hetmap matrix for report

import numpy as np
import pandas as pd
import seaborn as sns

# Create a DataFrame for the p-values
pval_matrix = pd.DataFrame({
    'Mangrove': [np.nan, 0.002, 0.001, 0.001],
    'Mixed forest': [0.002, np.nan, 0.626, 0.096],
    'Stunted hardwood': [0.001, 0.626, np.nan, 0.186],
    'Sawgrass': [0.001, 0.096, 0.186, np.nan]
}, index=['Mangrove', 'Mixed forest', 'Stunted hardwood', 'Sawgrass'])

# Display the p-value matrix
print(pval_matrix)
```

```python
# Create a mask for values above 0.05
mask = pval_matrix > 0.05

# Create the heatmap with masked non-significant values
sns.heatmap(pval_matrix, annot=True, cmap="viridis", linewidths=.5, fmt=".3f", 
            mask=mask, vmin=0, vmax=0.1, cbar_kws={'label': 'p-value'})

plt.title('Significant Pairwise ADONIS p-values (p < 0.05)', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load relative abundance data
abundance_df = pd.read_csv('class_freq.csv', index_col=0)

# Load metadata
metadata_df = pd.read_csv('metadata_grouped.tsv', delimiter='\t')

abundance_df = abundance_df.T

# Use sample id column for merging
abundance_df.reset_index(inplace=True)
abundance_df.rename(columns={'index': 'sample-id'}, inplace=True)


# Merge with metadata
merged_df = pd.merge(metadata_df, abundance_df, on='sample-id', how='inner')

import matplotlib.pyplot as plt

# List of abundance columns
abundance_columns = [
    'Crenarchaeota-Bathyarchaeia',
       'Proteobacteria-Gammaproteobacteria', 'Acidobacteriota-Acidobacteriae',
       'Crenarchaeota-Nitrososphaeria', 'Chloroflexi-Dehalococcoidia',
       'Proteobacteria-Alphaproteobacteria', 'Chloroflexi-Anaerolineae',
       'Sva0485-Undefined', 'Firmicutes-Clostridia', 'Firmicutes-Bacilli',
       'Archaea-Other', 'Bacteria-Other'
]

# Sort the DataFrame by Phasic community and depth cm
df_sorted = merged_df.sort_values(by=['Phasic_community', 'Depth_cm'])

# Calculate the average abundance for each phasic commm
average_abundance = df_sorted[abundance_columns].mean(axis=0)

# Sort the abundance columns by average abundance 
sorted_abundance_columns = average_abundance.sort_values(ascending=True).index.tolist()

# Plotting
fig, ax = plt.subplots(figsize=(14, 8))

# Create a stacked bar plot with sorted abundance columns where they are smallest to largest
bars = df_sorted[sorted_abundance_columns].plot(kind='bar', stacked=True, ax=ax, width=0.8, colormap='tab20')

# Set x to sample ID
ax.set_xticks(range(len(df_sorted)))
ax.set_xticklabels(df_sorted['Depth_cm'], rotation=45, ha='right')

# Add annotations for each Phasic_community
phasic_communities = df_sorted['Phasic_community'].unique()
positions = [df_sorted[df_sorted['Phasic_community'] == community].index for community in phasic_communities]
vertical_offset = -0.15

# Add annotations under the bars for each Phasic_community split out in quarters
for i, community_positions in enumerate(positions):
    start_pos = community_positions[0]
    end_pos = community_positions[-1]
    mid_pos = (start_pos + end_pos) / 2
    ax.text(mid_pos, vertical_offset, phasic_communities[i], ha='center', va='center', transform=ax.get_xaxis_transform(), fontsize=12)

# Reverse the order of the labels for the legend so from largest to smallest
handles, labels = ax.get_legend_handles_labels()
sorted_handles_labels = sorted(zip(handles, labels), key=lambda x: sorted_abundance_columns.index(x[1]))
handles, labels = zip(*sorted_handles_labels)

# Reverse the order of labels so that it is the same in the legend
ax.legend(handles[::-1], labels[::-1], loc='center left', bbox_to_anchor=(1.0, 0.5), title='Class')

# Customize plot
ax.set_xlabel('Depth (cm)')
ax.set_ylabel('Relative Abundance')
ax.set_title('')

plt.tight_layout()

# Show plot
plt.show()
```
